```python
data = [
    (1, "2024-01-05", "Tanisha  ", "Books", 2, 450),
    (2, "2024-01-10", "Rohit", "Electronics", 1, 1200),
    (3, "2024-01-12", "  Anjali", "Books", 3, 300),
    (4, "2024-01-15", "Vikas", "Clothing", 5, 700),
    (5, "2024-01-20", "tanisha", "Books", 1, 550),
    (6, "2024-01-22", None, "Electronics", 2, 2000)
]

columns = ["order_id", "order_date", "customer_name", "category", "quantity", "price"]

df = spark.createDataFrame(data, columns)
df.show()
df.printSchema()
```

```python
df.count()
df.show(3)
```

```python
from pyspark.sql.functions import col
df=df.withColumn("total_amount",col("quantity")*col("price"))
df.show()
```

```python
df2 = df.withColumn(
    "is_expensive",
    col("price") > 1000)
df2.show()
```

```python
from pyspark.sql.functions import lit
df=df.withColumn("source",lit("online")) 
df.show()
```

```python
from pyspark.sql.functions import when
df=df.withColumn(
    "priceflag",
    when(col("price") >=500, lit("High"))
    .otherwise(lit("low"))
)
df.show()
```

```python
df=df.withColumn("remarks",lit("None"))
```

```python
from pyspark.sql.functions import when,col
df=df.withColumn(
    "price_Category",
    when(col("price")>=1000 ,lit("Expensive"))
    .otherwise(lit("NORMAL"))
)
df.show()
```

```python
df.filter(col("category").isin("Books","Clothing")).show()
```

```python
df.filter(col("price").between(400,1000)).show()
```

```python
df.filter(col("customer_name").contains("Tanisha"))
```

```python
df.filter(col("customer_name").like("Tan%"))
```

```python
df.filter((col("category")=="Books") & (col("price")>400)).show()
```

```python
df.filter(col("customer_name").isNull()).show()
```

```python
df = df.withColumn("name_Status",
                   when(col("customer_name").isNull(), lit("Missing")).otherwise(lit("present")))
```

```python
df=df.withColumn(
    "order_size",
    when(col("quantity")>=3 ,lit("BULK"))
    .otherwise(lit("SMALL"))
)
df.show()
```

```python
df=df.withColumn("order_type",
                 when(col("price")>=1000,lit("High_Value"))
                .when(col("price")>=500 ,lit("Medium"))
                .otherwise(lit("LOW"))
                )
df.show()
```

```python
df = df.withColumn("customer_flag",
                   when(col("customer_name").isNull(),lit("unknown"))
                   .otherwise(lit("known")))
```

```python
from pyspark.sql.functions import concat ,lit
df=df.withColumn(
    "label",
    concat(lit("CAT_"),col("category"))
)
df.show()
```

```python
from pyspark.sql.functions import concat_ws
df=df.withColumn(
    "order_lavel",
    concat_ws("-",col("category"),col("order_id"))
)
```

```python
from pyspark.sql.functions import upper,split,trim
df=df.withColumn("customer_name_Clean",
                 upper(trim(col("customer_name"))))
df.show()
```

```python
from pyspark.sql.functions import to_date,col
df=df.withColumn(
    "order_Date_Clean",to_date(col("order_date"),"yyyy-MM-dd")
)
```

```python
from pyspark.sql.functions import current_date
df=df.withColumn("today",current_date())
df.show()
```

```python
from pyspark.sql.functions import current_timestamp
df=df.withColumn("run_time",current_timestamp())
```

```python
from pyspark.sql.functions import date_add,date_sub,col
df=df.withColumn("delivery_date",
                 date_add(col("order_date_clean"),7))
```

```python
from pyspark.sql.functions import year,month,dayofmonth
df=df.withColumn("order_year",year(col("order_date_Clean"))) \
    .withColumn("order_month",month(col("order_Date_clean")))
```

```python
from pyspark.sql.functions import to_date
df = df.withColumn(
    "order_date_clean",
    to_date(col("order_date"), "yyyy-MM-dd")
)
df.show()
```

```python
from pyspark.sql.functions import current_date,datediff
df=df.withColumn("days_old",
                                                                        datediff(current_date(),col("order_Date_Clean")))
df.show()
```
