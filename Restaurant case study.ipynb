```python
import pandas as pd 
sales_data = [
    ('A', '2025-01-01', 1),
    ('A', '2025-01-01', 2),
    ('A', '2025-01-07', 2),
    ('A', '2025-01-10', 3),
    ('A', '2025-01-11', 3),
    ('A', '2025-01-11', 3),
    ('B', '2025-01-01', 2),
    ('B', '2025-01-02', 2),
    ('B', '2025-01-04', 1),
    ('B', '2025-01-11', 1),
    ('B', '2025-01-16', 3),
    ('B', '2025-02-01', 3),
    ('C', '2025-01-01', 3),
    ('C', '2025-01-01', 3),
    ('C', '2025-01-07', 3)
]

sales = pd.DataFrame(
    sales_data,
    columns=["customer_id", "order_date", "product_id"]
)

sales["order_date"] = pd.to_datetime(sales["order_date"])

sales
```

```python
menu_data = [
    (1, 'biryani', 10),
    (2, 'paneer', 15),
    (3, 'dosai', 12)
]

menu = pd.DataFrame(
    menu_data,
    columns=["product_id", "product_name", "price"]
)

menu
```

```python
members_data = [
    ('A', '2025-01-07'),
    ('B', '2025-01-09')
]

members = pd.DataFrame(
    members_data,
    columns=["customer_id", "join_date"]
)

members["join_date"] = pd.to_datetime(members["join_date"])

members
```

- sales ["customer_id", "order_date", "product_id"]
- members=["customer_id", "join_date"]
- menu=["product_id", "product_name", "price"]


```python
# 1. Total amount spent by each customer 
sMm=sales.merge(menu , on="product_id" ,how="left")
Total_Amount=sMm.groupby("customer_id")["price"].sum()
Total_Amount
```

```python
# 2. Number of distinct visit days per customer
Distinct_customer=sales.groupby("customer_id") \
  .nunique("order_date")
Distinct_customer
```

```python
Distinct_customer=sales.groupby("customer_id")["order_date"].nunique()
Distinct_customer
```

```python
# 3. First item purchased by each customern
sales=sales.sort_values("order_date")
sales["f_item"]=sales.groupby("customer_id") \
    .cumcount() +1
sales
```

```python
 # 4. Most purchased item and count 
 most_purchased=(sales.groupby("product_id") \
                ["order_date"].count().idxmax())

most_purchased
```

```python

# 5. Most popular item per customer
sales_co = (
    sales.groupby(["customer_id", "product_id"])
         .size() \
    .reset_index(name="cnt"))

sales_co["rank"]= sales_co.groupby("customer_id")["cnt"] \
                        .rank(method="min" , ascending=False)
most_popular_item = sales_co[sales_co["rank"] == 1]
most_popular_item
```

```python
 # 6. First item after becoming a member
sales_member=sales.merge(members , on="customer_id",how="left")
sales_member=sales_member.sort_values("order_date")
after_join = sales_member[sales_member["order_date"] >= sales_member["join_date"]]
first_item=(after_join.groupby("customer_id").cumcount() + 1)
after_join[first_item == 1]
```

```python
# 7. Last item before becoming a member
sales_member = sales.merge(members, on="customer_id" , how="left")
before_join = sales_member[sales_member["order_date"] <= sales_member["join_date"]]
before_join=before_join.sort_values("order_date" , ascending=False)
last_item=before_join.groupby("customer_id").cumcount() +1
before_join[last_item == 1]
```

```python

 # 8. Items and amount before becoming a member ***sumation of price 
 item_Amount=sales.merge(members ,on="customer_id" , how="left" ) \
                    .merge(menu,on="product_id",how="left")
before=item_Amount[item_Amount["order_date"]<item_Amount["join_date"]]
before=(before.groupby("customer_id")["price"].sum().reset_index(name="total_Amount"))
before
```

```python
# 9. Loyalty points: 2x for biryani, 1x for others 
import numpy as np
import pandas as p
sales_menu=sales.merge(menu,on="product_id",how="left")
sales_menu["points"]=np.where(
    sales_menu["product_name"] == "biryani",
    sales_menu["price"]*2,
    sales_menu["price"]
)
sales_menu
```

```python
# 10. Points during first 7 days after joining
smm=sales.merge(menu,on="product_id",how="left") \
        .merge(members,on="customer_id",how="left")
smm7=smm[(smm["order_date"]>=smm["join_date"]) & (smm["order_date"]<=smm["join_date"] + pd.to_timedelta(7,unit = "D")) ]
smm7["points"]=np.where(smm7["product_name"]=="biryani",
                        smm7["price"] *2,
                        smm7["price"])
smm7
```

```python
 # 11. Total spent on biryani
 sales_MM=sales.merge(menu,on="product_id",how="left") \
        .merge(members,on="customer_id",how="left")
sales_biryani=sales_MM[sales_MM["product_name"]=="biryani"]["price"].sum()
display(sales_biryani)
```

```python
 # 12. Customer with most dosai orders 
sales_menu = sales.merge(menu, on="product_id", how="left")
dosai_order=sales_menu[sales_menu["product_name"]=="dosai"]

dosai_count =(dosai_order.groupby("customer_id").size())
most_dosai_customer = dosai_count.idxmax()
most_dosai_customer
```

```python
# 13. Average spend per visit 
sales_sum=sales_menu.groupby(["order_date","customer_id"])["price"].sum()
sales_avg=sales_sum.mean()
sales_avg
```

```python

# 14. Day with most orders in Jan 2025
jan_2025 = sales[
    (sales["order_date"].dt.year == 2025) &
    (sales["order_date"].dt.month == 1)
]
day_most=jan_2025.groupby("order_date").size().idxmax()
day_most
```

```python
 # 15. Customer who spent the least 

 spent_least=sales_menu.groupby("customer_id")["price"].sum().idxmin()
 spent_least
```

```python
# 16. Date with most money spent
money=sales_menu.groupby("order_date")["price"].sum().idxmax()
money
```
