```python
bike = pd.DataFrame({
    "id": [1,2,3,4],
    "category": ["mountain bike", "road bike", "hybrid", "electric"],
    "price_per_hour": [10,12,8,15],
    "price_per_day": [50,60,40,75],
    "status": ["available","available","rented","available"]
})
```

```python
rental = pd.DataFrame({
    "id": [1,2,3,4],
    "customer_id": [1,1,2,3],
    "bike_id": [1,2,3,4],
    "start_timestamp": pd.to_datetime([
        "2023-01-05","2023-02-05","2023-02-10","2023-03-15"
    ]),
    "total_paid": [50,60,40,75]
})
```

```python
membership_type = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Basic Monthly", "Basic Annual", "Premium Monthly"],
    "description": [
        "Unlimited rides with non-electric bikes. Renews monthly.",
        "Unlimited rides with non-electric bikes. Renews annually.",
        "Unlimited rides with all bikes. Renews monthly."
    ],
    "price": [100.00, 500.00, 200.00]
})
```

```python
membership = pd.DataFrame({
    "id": [1,2,3,4,5,6],
    "membership_type_id": [2,1,3,1,2,3],
    "customer_id": [3,2,4,1,2,3],
    "start_date": pd.to_datetime([
        "2023-08-01","2023-08-01","2023-08-01",
        "2023-09-01","2023-09-01","2023-09-01"
    ]),
    "end_date": pd.to_datetime([
        "2023-08-31","2023-08-31","2023-08-31",
        "2023-09-30","2023-09-30","2023-09-30"
    ]),
    "total_paid": [500.00, 100.00, 200.00, 100.00, 500.00, 200.00]
})
```

```python
# Emily would like to know how many bikes the shop owns by category.
# Display the category name and the number of bikes the shop owns in each category (call this column number_of_bikes).
# Show only the categories where the number of bikes is greater than 2.
import numpy as np
total_bike=bike.groupby("category")["id"].count().reset_index(name="number_of_bike")
total_bike=total_bike[total_bike["number_of_bike"]>2]
total_bike
```

```python
# Emily needs a list of customer names with the total number of memberships purchased by each.
# For each customer, display the customer's name and the count of memberships purchased (call this column membership_count).
# Sort the results by membership_count, starting with the customer who has purchased the highest number of memberships.
# Keep in mind that some customers may not have purchased any memberships yet.
# In such a situation, display 0 for the membership_count.
customer_rental=customer.merge(membership,left_on="id",right_on="customer_id",how="left")
customer_list=customer_rental.groupby("name")["membership_type_id"].count() \
            .reset_index(name="membership_count")\
                   .sort_values("membership_count", ascending=False)
customer_list
```

```python
# Emily is working on a special offer for the winter months.
# For each bike, display its ID, category, old price per hour (call this column old_price_per_hour), discounted price per hour (call it new_price_per_hour), old price per day (call it old_price_per_day), and discounted price per day (call it new_price_per_day).

# Electric bikes should have a 10% discount for hourly rentals and a 20% discount for daily rentals.
# Mountain bikes should have a 20% discount for hourly rentals and a 50% discount for daily rentals.
# All other bikes should have a 50% discount for all types of rentals.
# Round the new prices to 2 decimal digits.
bike["new_price_per_hour"]=np.where(
    bike["category"]=="electric",
    bike["price_per_hour"]*0.90,
    np.where(
        bike["category"] == "mountain bike",
        bike["price_per_hour"] * 0.80,
    bike["price_per_hour"]*0.5
)
)
bike["new_price_per_day"]=np.where(
    bike["category"]=="electric",
    bike["price_per_day"] *0.80,
    np.where(
        bike["category"]=="mountain bike",
        bike["price_per_day"] * 0.50,
    bike["price_per_hour"]*0.5
    )
)
bike=bike.rename(columns={
    "price_per_hour":"old_price_per_hour",
    "price_per_day":"old_price_per_day"
})
bike[["id","category","old_price_per_hour","old_price_per_day","new_price_per_hour","new_price_per_day"]]
```

```python
# Emily is looking for counts of the rented bikes and of the available bikes in each category.
# Display the number of available bikes (call this column available_bikes_count) and the number of rented bikes (call this column rented_bikes_count) by bike category.
bike4=bike[["category","status"]]
groupbike=bike4.groupby(["category","status"]).size().unstack(fill_value=0)
bike_name=groupbike.rename(columns={
    "available":"available_bikes_count",
    "rented":"rented_bikes_count"
})
bike_name
```

```python
# Emily has asked you to get the total revenue from memberships for each combination of year, month, and membership type.
# Display the year, the month, the name of the membership type (call this column membership_type_name), and the total revenue (call this column total_revenue) for every combination of year, month, and membership type.
# Sort the results by year, month, and name of membership type.
# Membership se year & month nikalo
# Membership ko membership_type se join karo
# Group by:
# year
# month
# membership type name
# Sum(total_paid)
# Sort

membership["year"]=membership["start_date"].dt.year
membership["month"]=membership["start_date"].dt.month
member_type=membership.merge(membership_type,left_on="membership_type_id",right_on="id",how="left")
member_type=member_type.groupby(by=["year","month","name"])["total_paid"].sum().reset_index(name="total_revenue").rename(columns={"name": "membership_type_name"}).sort_values(by=["year", "month", "membership_type_name"])
member_type
```

```python
# Emily wants to segment customers based on the number of rentals and see the count of customers in each segment.
# Categorize customers based on their rental history as follows:
# Customers who have had more than 10 rentals are categorized as 'more than 10'.
# Customers who have had 5 to 10 rentals (inclusive) are categorized as 'between 5 and 10'.
# Customers who have had fewer than 5 rentals should be categorized as 'fewer than 5'.
# Calculate the number of customers in each category.
# Display two columns: rental_count_category and customer_count.
import numpy as np
rental_bikes=rental.groupby("customer_id")["bike_id"].count().reset_index(name="rental_count")
rental_bikes["rental_count_category"]=np.where(
    rental_bikes["rental_count"] >10,"more than 10",
    np.where(rental_bikes["rental_count"]>=5 ,"between 5 and 10",
             "fewer than 5")
)
result = rental_bikes.groupby("rental_count_category")["customer_id"].count().reset_index(name="customer_count")
result
```

```python
import pandas as pd
import numpy as np

# 1️⃣ Year & month nikaalo
membership["year"]  = membership["start_date"].dt.year
membership["month"] = membership["start_date"].dt.month

# 2️⃣ Sirf 2023 ka data
membership_2023 = membership[membership["year"] == 2023]

# 3️⃣ membership ↔ membership_type join
full = membership_2023.merge(
    membership_type,
    left_on="membership_type_id",
    right_on="id",
    how="left"
)

# 4️⃣ Detail: membership_type + month
detail = (
    full.groupby(["name", "month"])["total_paid"]
        .sum()
        .reset_index(name="total_revenue")
)

# 5️⃣ Subtotal by membership type (month = NA)
type_total = (
    full.groupby("name")["total_paid"]
        .sum()
        .reset_index(name="total_revenue")
)
type_total["month"] = pd.NA

# 6️⃣ Subtotal by month (membership_type = NA)
month_total = (
    full.groupby("month")["total_paid"]
        .sum()
        .reset_index(name="total_revenue")
)
month_total["name"] = pd.NA

# 7️⃣ Grand total (both NA)
grand_total = pd.DataFrame({
    "name": [pd.NA],
    "month": [pd.NA],
    "total_revenue": [full["total_paid"].sum()]
})

# 8️⃣ Sab combine
final = pd.concat(
    [detail, type_total, month_total, grand_total],
    ignore_index=True
)

# 9️⃣ Column rename + AND sorting
final = (
    final.rename(columns={"name": "membership_type_name"})
         .sort_values(
             by=["membership_type_name", "month"],
             ascending=[True, True],
             na_position="last"
         )
         .reset_index(drop=True)
)

final
```
